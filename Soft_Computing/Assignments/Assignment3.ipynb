# Develop a product recommendation system using fuzzy logic to suggest the appropriate electronics gadgets on an e-commerce app based on the rating of the user for the similar items.

```python
import numpy as np
import pandas as pd
import math
```

```python
# Define the fuzzy membership function for the "rating" variable
def rating_membership(value):
    return value/5
```

```python
# Load the sample data
data = {
    'item': ['Laptop', 'SmartPhone', 'AirConditioner', 'Television'],
    'user1': [3.4, 5, 3.1, 1.9],
    'user2': [1.8, 4.5, 2.7, 3.7],
    'user3': [1.5, 2.6, 4.7, 3.9],
    'user4': [3.3, 4.1, 4.1, 2.2]
}
df = pd.DataFrame(data)
# Calculate the membership degrees for each user and item
for i in range(1, 5):
    user_column = 'user' + str(i)
    like_values = []
    for j in range(len(df)):
        like_values.append(rating_membership(df[user_column][j]))
    df['like_user' + str(i)] = like_values

# Find the recommended items for each user
for i in range(1, 5):
    user_like_column = 'like_user' + str(i)
    recommendations = []
    for j in range(len(df)):
        item_like_value = df[user_like_column][j]
        if item_like_value > 0.6:
            recommendations.append(df['item'][j])
    print("User " + str(i) + " recommendations: " + str(recommendations))
```
