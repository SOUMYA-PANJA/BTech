# Implement a program that defines and evaluates fuzzy membership functions for a given set of inputs and outputs the degree of membership for each input.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fuzzy_union(set1, set2):
    """
    Perform union operation on two fuzzy sets
    """
    return np.maximum(set1, set2)
def fuzzy_intersection(set1, set2):
    """
    Perform intersection operation on two fuzzy sets
    """
    return np.minimum(set1, set2)
def fuzzy_complement(set1):
    """
    Perform complement operation on a fuzzy set
    """
    return 1 - set1
```

```python
set1 = np.array([0.3, 0.6, 0.4])
set2 = np.array([0.7, 0.4, 0.8])

union = fuzzy_union(set1, set2)
intersection = fuzzy_intersection(set1, set2)
complement = fuzzy_complement(set1)
print(f"Union: {union}")
print(f"Intersection: {intersection}")
print(f"Complement: {complement}")
```
