# Develop a loan approval system using fuzzy logic to assess the loan application based on the credit score and income of the applicant. Also plot the membership function operation between the credit score and loan acceptation/rejection.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoidInc(x):
    return 1/(1+np.exp(-0.2*(x - 5e5)))

def sigmoidCS(x):
    return 1/(1+np.exp(-0.3*(x - 550)))
```

```python
def loan(X, Y):
    approvals = []
    for i in range(len(X)):
        if(max(sigmoidInc(X[i]), sigmoidCS(Y[i])) > 0.5):
            approvals.append(1)
        else:
            approvals.append(0)
    return approvals
```

```python
incomes = [500000, 100000, 750000, 250000, 450000]
credit_scores = [670, 500, 400, 200, 690]
results = loan(incomes, credit_scores)
```

```python
plt.scatter(credit_scores, results, c = results)
```
