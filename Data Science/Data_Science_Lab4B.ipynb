```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# sns.set_style("whitegrid")
plt.figure(figsize = (10, 10))
sns.set_style("darkgrid")
data = np.random.normal(size = (20, 10)) + np.arange(10)/ 2
sns.boxplot(data = data)
```
