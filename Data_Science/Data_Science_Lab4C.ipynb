```python
fig = px.histogram(data, marginal = "box", x = "Calories", title = "Calories Distribution")
fig.update_layout(bargap = 0.2)
# fig.show()
```

```python
plt.figure(figsize = (20, 10))
sns.histplot(data = data, x = "Calories")
plt.xlabel("Calories")
plt.xticks(rotation = 90)
plt.ylabel("Count")
```
