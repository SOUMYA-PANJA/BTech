```python
data = {
    "director":
            [
            "Kirsten Johnson","Mike Flanagan","Andy Devonshire","Bruno Garotti","Adam Salky","Scott Stewart"
            ],
    "Total Runtime":
            [
            144, 120, 114, 92, 90, 91
            ],
    "Genre":
            [
            "Adventure","Comedy","Documentary","Crime","Sports","Mystery"
            ]
}
df2 = pd.DataFrame(data)
df2
```

```python
above_100_episodes = df2[df2['Total Runtime']>100]
```

```python
above_100_episodes
```

```python
above_100 = df2[df2['Total Runtime'].map(lambda x:x>100)]
above_100
```

```python
above_100_check = list()
for i in df2['Total Runtime']:
  if i > 100:
    above_100_check.append(True)
  else:
    above_100_check.append(False)

ser = pd.Series(above_100_check)
above_100_movies = df2[ser]
above_100_movies
```

```python
above_100_adventure = df2[(df2['Total Runtime']>100) & (df2['Genre']=='Adventure')]
above_100_adventure
```

```python
less_than_50_drama = df2[df2['Total Runtime'].map(lambda x:x<=50) & df2['Genre'].map(lambda movie:movie == "Sports")]
less_than_50_drama
```

```python
maximum_time = df2[df2['Total Runtime']==df2['Total Runtime'].max()]
maximum_time
```

```python
minimum_time = df2[df2['Total Runtime']==df2['Total Runtime'].min()]
minimum_time
```

```python
latest_release_year = df[df['release_year'] == df['release_year'].max()]
latest_release_year.max()
```

```python
old_release_year = df[df['release_year'] == df['release_year'].min()]
old_release_year
```
