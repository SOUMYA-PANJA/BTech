```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (15, 12))
ax = plt.axes()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid("on")
plt.show()
```

```python
x = np.linspace(0, 15, 150)
plt.plot(np.sin(x), label = "Sine Wave", color = "g", marker = "H")
plt.plot(np.cos(x), label = "Cosine Wave", color = "orange", marker = "D")
plt.legend(loc = "upper right")
plt.savefig('trigono.png')
plt.show()
```

```python
x = np.linspace(0, 15, 150)
plt.plot(np.sin(x), label = "Sine Wave", color = "violet", marker = "^")
plt.plot(np.cos(x), label = "Cosine Wave", color = "aqua", marker = "+")
plt.legend(loc = "upper right")
plt.title("Trignometry Function")
plt.xlabel("X")
plt.ylabel("sin(x), cos(x)")
plt.show()
```

```python
x = [12, 45, 23, 10, 30]
y = [20, 17, 34, 3, 10]
plt.scatter(x, y, color = "red", marker = "+", label = "X to Y")
plt.scatter(y, x, color = "blue", marker = "*", label = "Y to X")
plt.legend()
```
