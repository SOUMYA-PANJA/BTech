```python
from sklearn.datasets import make_blobs
clstr_centers = 10
train_samples = 50000
test_samples = 100
spread = 0.5
rndm_state = 3
bat_size = 32
```

```python
import tensorflow as tf

print(tf.__version__)
```

```python
X, Y = make_blobs(centers = clstr_centers, n_samples = train_samples + test_samples, random_state = rndm_state, cluster_std = spread)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = train_samples)
```

```python
import numpy as np
np.set_printoptions(linewidth=320)
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Dense(32, activation = tf.nn.relu),
                                    tf.keras.layers.Dense(10, activation = tf.nn.softmax)])
```

```python
model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs = 100, batch_size = bat_size)
```

```python
model.evaluate(X_test, Y_test)
```
