# Implement a program that performs union, intersection, and complement operations on fuzzy sets from user input set element values and outputs the results with appropriate map plotting.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fuzzy_union(set1, set2):
    """
    Calculates the union of two fuzzy sets.
    """
    union = []
    for i in range(min(len(set1), len(set2))):
        union.append(max(set1[i], set2[i]))
    return union
```

```python
def fuzzy_intersection(set1, set2):
    """
    Calculates the intersection of two fuzzy sets.
    """ 
    intersection = []
    for i in range(min(len(set1), len(set2))):
        intersection.append(min(set1[i], set2[i]))
    return intersection
```

```python
def fuzzy_complement(set1):
    """
    Calculates the complement of a fuzzy set.
    """
    complement = []
    for i in range(len(set1)):
        complement.append(1 - set1[i])
    return complement
```

```python
def plot_sets(setA):
    """
    Plot the membership functions of fuzzy sets.
    """
    x = np.arange(len(setA))
    plt.stem(x, setA)
    plt.xlabel('Universe of Discourse')
    plt.ylabel('Membership Degree')
    plt.ylim(0, 1)
    plt.show()
```

```python
setA = []
setB = []

n = int(input("Enter number of elements in set1: "))
print("Enter elements of set 1:")
for i in range(n):
    setA.append(float(input()))

n = int(input("Enter number of elements in set2: "))
print("Enter elements of set 2:")
for i in range(n):
    setB.append(float(input()))

union = fuzzy_union(setA, setB)
intersection = fuzzy_intersection(setA, setB)
complement = fuzzy_complement(setA)
```

```python
plot_sets(setA)
plot_sets(setB)
```

```python
plot_sets(union)
```

```python
plot_sets(intersection)
```

```python
plot_sets(complement)
```
